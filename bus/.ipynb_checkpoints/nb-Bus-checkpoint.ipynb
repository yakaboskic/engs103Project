```python
def city_rand():
    rand = random.random()
    start = 0
    end = stats.Lebanon
    if rand <= end:
        return 'Lebanon'
    start = end
    end = start + stats.Sachem
    if rand > start and rand <= end:
        return 'Sachem'
    start = end
    end = start + stats['West Lebanon']
    if rand > start and rand <= end:
        return 'West Lebanon'
    start = end
    end = start + stats['White River Junction']
    if rand > start and rand <= end:
        return 'White River Junction'
    start = end
    end = start + stats.Norwich
    if rand > start and rand <= end:
        return 'Norwich'
    start = end
    end = start + stats['Enfield']
    if rand > start and rand <= end:
        return 'Enfield'
    else:
        return 'Issue'
```

```python
df_dist = pd.read_csv('/home/yakaboskic/courses/engs103Project/data/UpperValleyDists.csv')
df_dist = df_dist.set_index('Unnamed: 0')
df_dist = df_dist.fillna(0)
df_dist
```

```python
def get_tsp_trip(stops):
    start = 'Hanover'
    end = 'Hanover'
    distance_results = []
    stops = set(stops)
    for perm in itertools.permutations(stops, r=len(stops)):
        tour = list(perm)
        tour.insert(0, 'Hanover')
        tour.append('Hanover')
        distance = 0
        start = 'Hanover'
        for stop in tour:
            distance = distance + df_dist.loc[start, stop]
            start = stop
        distance_results.append([distance, tour])
    return sorted(distance_results)[0]
```

```python
_stops = copy.deepcopy(stops)
bus_route = get_tsp_trip(_stops)
bus_route
```

```python
BUS_MPH = 25
BUS_CAPACITY_MAX = 30
route_time = bus_route[0]/BUS_MPH
matrix = []
for capacity in range(1, BUS_CAPACITY_MAX):
    serviceable_lambda = route_time*capacity
    _avg_wq = route_time/2
    _avg_lq = _avg_wq*serviceable_lambda
    matrix.append((capacity, serviceable_lambda, _avg_wq, _avg_lq))

df_servicable = pd.DataFrame(matrix, columns=['Bus Capacity', 'Serviceable Lambda', 'Avg. Wq', 'Avg. Lq'])
df_servicable['Expected Customers Serviced (Nightly)'] = df_servicable['Serviceable Lambda']*4
df_servicable = df_servicable.set_index('Bus Capacity')
df_servicable
```

```python
BUS_MPH = 25
BUS_CAPACITY_MAX = 30
route_time = bus_route[0]/BUS_MPH
matrix = []
for lamb in np.linspace(.5, 5, 15):
    #serviceable_lambda = route_time*capacity
    _avg_wq = route_time/2
    _avg_lq = _avg_wq*lamb
    matrix.append((capacity, lamb, _avg_wq, _avg_lq))

df = pd.DataFrame(matrix, columns=['Bus Capacity', 'Lambda', 'Avg. Wq', 'Avg. Lq'])
df['Expected Customers Serviced (Nightly)'] = df['Lambda']*4
df
```

```python
df.to_csv('cost_bus.csv')
```

```python
BUS_MPG = 10
COST_GALLON_GAS = 3
BUS_DRIVER_SALARY = 10
BUS_COST = 0
EXPECTED_MILEAGE_BEFORE_REPLACE = 100000
DEPRECIATION = BUS_COST/EXPECTED_MILEAGE_BEFORE_REPLACE
COST_PER_MILE = COST_GALLON_GAS/BUS_MPG + DEPRECIATION
hours_per_day = 4
days = 2*52
total_hours = hours_per_day*days
total_salary = BUS_DRIVER_SALARY*total_hours

nightly_cost = bus_route[0]*COST_PER_MILE*(4/route_time)
total_cost = nightly_cost*total_hours + total_salary + BUS_COST
total_cost

df_servicable['Cost per student served'] = total_cost/(df_servicable['Expected Customers Serviced (Nightly)']*(2*52))
ax = df_servicable.plot.line(x=0, y=-1, title='Serviceable Labmbda vs. Cost per Student Serviced')
ax.set_ylabel('Cost ($)')
```
